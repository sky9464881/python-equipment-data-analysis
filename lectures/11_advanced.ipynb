```python
#OEE (Overall Equipment Effectiveness)
```

```python
# 가동률*성능률*양품률
```

```python
# 가동률 = 실제 가동시간/ 계획 가동시간
# 성능률 = ( 실제 생산량 * 이상 사이클 타임) /실제 가동시간
# 양품률 = 양품수량/ 총생산량
```

```python
# OEE 구해서 50% 미만 -> 현재의 설비 효율 개선하자.
# 85 이상 -> 설비 증설 검토
```

```python
### 가동률이 낮다 -> 고장/ 정지시간을 줄이기 위한 방법을 찾아야 한다.
### 성능률이 낮다 -> 속도 개선, 교체 시간 단축 위한 방법을 찾아야 한다.
### 양품률이 낮다 -> 품질 개선, 재작업 감소하는 방법을 찾아야 한다.
```

```python
### 벤치마킹 가능한다 -> 설비간 비교를 통해서, 우수 설비를 찾아낸다.
```

```python
### 공장 간 비교가 가능하다. 우수공장은 Best paractice가 된다.
```

```python
#OEE 계산 결과
# 85% 이상: 월클 수준
# 70~84% : 양호
# 60~69% : 보통 -> 개선 과제 도출
# 60% 미만 : 개선필요, 긴급 조치
```
