```python
# 설비의 온도를 주면 현재상황을 리턴하는 함수 (critical, warning, normal, low)
```

```python
def get_temp_status(temp) :
    if temp >= 200 :
        return 'critical'
    elif temp >= 190 :
        return 'warning'
    elif temp >= 180 :
        return 'noraml'
    else :
        return 'low'
    
```

```python
#온도 리스트를 주면 이 데이터의 최소값, 최대값, 평균을 리턴하는 함수
```

```python
def get_statistics(temp_list) :  ## 파이썬은 여러개를 리턴 가능
    """ 데이터 기본 통계량 리턴"""
    min_value = min(temp_list)
    max_value = max(temp_list)
    mean_value = sum(temp_list)/len(temp_list)
    print(f"min: {min_value}, max: {max_value}, mean: {mean_value}")
```

```python
get_statistics([185.5,190,147,170])
```

```python
#온도를 점검할 때(WARNING, NORMAL) 온도와 임계값을 줄거다.
# 임계값은 기본으로 190을 사용하되, 변경 가능토록 하고 싶다.

def check_temp(temp, threshold = 190 ) :   #default값을 정해 줄 수 있음
    if temp > threshold :
        return 'WARNING'
    else :
        return 'NORMAL'
```
