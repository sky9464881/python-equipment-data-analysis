```python
machine_name = "CNC Machine"
current_temp = 185.5
target_temp = 180.0
production = 1500
efficiency = 0.856
```

```python
temp_diff = current_temp - target_temp

print(f"[설비 상태 보고서]")
print(f"설비명: {machine_name}")
print(f"현재 온도: {current_temp:.1f}°C")
print(f"목표 온도: {target_temp:.1f}°C")
print(f"온도 편차: {temp_diff:+.1f}°C")
print(f"생산량: {production:,}개")
print(f"효율: {efficiency:.1%}")
```
