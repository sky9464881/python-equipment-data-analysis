```python
df_production['불량률'] = df_production['defect_quantity'] / df_production['actual_quantity'] * 100
```

```python
#불량률이 5프로 미만이면 '우수'
#그렇지 않으면 '주의'라고 등급 컬럼을 만들자
```

```python
df_production
```

```python
### np.select()
```

```python
#3이상 조건으로 분류할 때
```

```python
# 불량륭이 3보다 작으면 S등급
#7:A, 10:B, 그렇지 않으면 C등급
```

```python
conditions = [df_production['불량률']<3,df_production['불량률']<7,df_production['불량률']<10]
```

```python
choices = ['S등급','A등급','B등급']
```

```python
default ='C등급'
```

```python
np.select(conditions,choices,default=default)
```

```python
df_production['4등급'] = np.select(conditions,choices,default=default)
```

```python
df_production
```

```python
###판다스의 apply()로 사용하는 방법
```

```python
df_production['불량률']
```

```python
def get_grade(rate) : 
    if rate<5 :
        return '우수'
    else :
        return '주의'
```

```python
get_grade(4.938272)
```

```python
df_production['불량률'].apply( get_grade )  # 판다스 의 방식
```

```python
def get_grade_e(rate) :
    if rate < 3 :
        return 'S등급'
    elif rate < 7 :
        return 'A등급'
    elif rate < 10 : 
        return 'B등급'
    else :
        return 'C등급'
```

```python
df_production['불량률'].apply( get_grade_e )  # 판다스 의 방식
```
