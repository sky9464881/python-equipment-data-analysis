```python
production_line = [
    {"position": 1, "product_id": "P001", "status": "OK", "temperature": 185.5},
    {"position": 2, "product_id": "P002", "status": "OK", "temperature": 186.2},
    {"position": 3, "product_id": "P003", "status": "NG", "temperature": 192.5},
    {"position": 4, "product_id": "P004", "status": "OK", "temperature": 184.8},
    {"position": 5, "product_id": "P005", "status": "CRITICAL", "temperature": 205.0},
    {"position": 6, "product_id": "P006", "status": "OK", "temperature": 185.0},
]
```

```python
OK=0
NG=0
for data in production_line :
    print(f"[위치 {data['position']}] {data['product_id']}: {data['status']} ({data['temperature']})", end='')
    if data['temperature']>=190 :
        print(' ⚠️ 고온 경고')
    else: print()
    if data['status']=='NG' or data['status']=='CRITICAL' :
        NG+=1
    if data['status']== 'CRITICAL' :
        print('!!! 치명적 불량 발견 - 라인 긴급정지 !!!')
        break
    if data['status']== 'OK' :
        OK+=1
print()
print("------검사결과---------")
print(f"검사 제품 수 : {data['position']} 개")
print(f"합격: {OK}개")
print(f"불량: {NG}개")
if data['status']== 'CRITICAL' :
    print(f"라인상태 : 긴급정지")
else :
    print(f"라인상태 : 정상")
    
```
