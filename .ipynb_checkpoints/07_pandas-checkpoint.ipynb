```python
equipment_data = {       #key 값이 컬럼
    'equipment_id': ['INJ-001', 'INJ-002', 'PRESS-001', 'PRESS-002', 'ASM-001'],
    'equipment_name': ['사출기 1호기', '사출기 2호기', '프레스 1호기', '프레스 2호기', '조립라인 1호기'],
    'equipment_type': ['사출기', '사출기', '프레스', '프레스', '조립라인'],
    'location': ['A동 1라인', 'A동 1라인', 'A동 2라인', 'A동 2라인', 'B동 1라인'],
    'rated_capacity': [150, 150, 200, 200, 100],
    'status': ['ACTIVE', 'ACTIVE', 'ACTIVE', 'ACTIVE', 'ACTIVE']
}
```

```python
equipment_data
```

```python
#판다스에서 2차원을 DataFrame이라고 부른다
```

```python
pd.DataFrame(data=equipment_data) #index,values,columns (구분 잘하기)
```

```python
df=pd.DataFrame(data=equipment_data)
```

```python
df
```

```python
df=df.set_index('equipment_id')
```

```python
#dataframe에서 데이터를 가져오는 방법
# 3가지 방법
# 1. 컬럼 데이터 가져오기
# 2. .loc[,] 로 가져오기
# 3. .iloc[,] 로 가져오기
```

```python
df
```

```python
df['equipment_name'] #인덱스와 함께 equipment_name이 가져와짐 (바로 대괄호는 colmn을 선택할때)
```

```python
df[['equipment_name','rated_capacity']]
```

```python
df[['equipment_name' ,'location', 'rated_capacity']]  # [:]은 안됨
```

```python
df
```

```python
df.loc['INJ-002',:] #loc는 index와 colmn으로 데이터 가져오는 방식
```

```python
df.loc['INJ-002':'PRESS-001',:]
```

```python
df.loc[:,'equipment_type':'rated_capacity']
```

```python
df[['equipment_type','location']].loc['INJ-002':'PRESS-002',:]
```

```python
df
```

```python
df.iloc[1:4,1:4] #iloc는 숫자로 인덱스 판별
```
