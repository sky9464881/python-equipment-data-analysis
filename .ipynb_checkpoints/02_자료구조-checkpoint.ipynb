```python
temperatures = [185.5, 192.3, 178.2, 195.1, 182.4, 188.7, 201.2, 176.8]
```

```python
print(f'최고온도: {max(temperatures)}') 
print(f'최저온도: {min(temperatures)}') 
print(f'평균온도: {sum(temperatures)/len(temperatures)}')
print(f'정렬된 온도: {sorted(temperatures)}')
temperatures.sort();
print(f'190도 이상: {len(temperatures[6:])}개')
```

```python
production = [120, 135, 128, 142, 138, 145, 132, 140, 136, 150]
```

```python
print(f'첫 5개: {production[:6]}')
print(f'마지막 3개: {production[-3:]}')
print(f'홀수 인덱스: {production[1::2]}')
print(f'역순: {production[::-1]}')
print(f'중간데이터: {production[2:-2]}')
```

```python
shift_A = [3, 5, 2, 4]
shift_B = [1, 2, 3, 2]
```

```python
print(f'연결: {shift_A+shift_B}')
print(f'반복: {shift_A*3}')
print(f'총 불량: {sum(shift_A+shift_B)}')
print(f'최대: {max(shift_A+shift_B)}')
print(f'최소: {min(shift_A+shift_B)}')
```

```python
sensor_data = [
    [185.5, 186.2, 184.8, 185.0, 186.1],
    [120.3, 121.5, 119.8, 120.1, 120.9],
    [45.2, 44.8, 45.5, 45.1, 44.9]
]
```

```python
print(f'센서 1 평균: {sum(sensor_data[0])/len(sensor_data[0])}')
print(f'센서 2 최댓값: {max(sensor_data[1])}')
print(f'센서 3 최솟값: {min(sensor_data[1])}')
print(f'센서 3 첫 3개: {sensor_data[2][:3]}')
print(f'모든 센서의 첫번째 값: {[sensor_data[0][0],sensor_data[1][0],sensor_data[2][0]]}')
```
